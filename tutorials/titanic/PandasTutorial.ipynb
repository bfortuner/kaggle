#### Cleaning The Data

```python
# Create new column where all values = 4
df['Gender'] = 4
df.head(2)
```

```python
# Gender equals M or F
df['Gender'] = df['Sex'].map( lambda x: x[0].upper() )
df.head(2)
```

```python
# Let's use ints instead
df['Gender'] = df['Sex'].map( {'female': 0, 'male': 1} ).astype(int)
df.head(2)
```

```python
print df.info()
# Get distinct values
print "Unique Embarked Values: " + str(df['Embarked'].unique())
print df.head(2)[['Age','Gender','Embarked']]

#Drop rows with null Embarked
#DataFrame is modified!
df.dropna()
print "Count all Rows: " + str(df.count())
print "Count Embarked which should be the same: " + str(df['Embarked'].count())


#Modify df Embarked Column (you would have drop the null rows first)
#df['Embarked'] = df['Embarked'].map( {'S': 0, 'C': 1, 'Q':2} ).astype(int)
print df.head(2)[['Age','Gender','Embarked']]

```

```python
# Create a lookup table for Ages by passenger class
median_ages = np.zeros((2,3))
print "Initialized Lookup Table"

#Rows equal Gender
#Columns represent unique Pclass values
print median_ages
```

```python
#How many distinct classes are there?
print "Unique pClass: " + str(df['Pclass'].unique())

# Populate the Lookup Table w Query!
for i in range(0, 2):   #For each Gender (row)
    for j in range(0, 3):     #For each pClass (col) in our lookup table
        median_ages[i,j] = df[(df['Gender'] == i) & \
                              (df['Pclass'] == j+1)]['Age'].dropna().median()
median_ages
```

```python
#Create a new column for normalized Age to be safe
df['AgeFill'] = df['Age']
df.head(2)
```

```python
#Inspect missing values
df[ df['Age'].isnull() ][['Gender','Pclass','Age','AgeFill']].head(10)
```

```python
# Set Age Fill Null Cells
for i in range(0, 2):
    for j in range(0, 3):
        #df.loc finds all occurences that match your query
        df.loc[ (df.Age.isnull()) & (df.Gender == i) & (df.Pclass == j+1),\
                'AgeFill'] = median_ages[i,j]

df[ df['Age'].isnull() ][['Gender','Pclass','Age','AgeFill']].head(10)
```

```python
# Create new column that tells us if Age is null
df['AgeIsNull'] = pd.isnull(df.Age).astype(int)
```

```python
# Big Time Aggregations!
# Define the aggregation calculations
aggregations = {
    'Survived': { # work on the "duration" column
        'total_passengers': 'count',  # get the sum, and call this result 'total_duration'
        'total_survived': 'sum', # get mean, call result 'average_duration'
        'mean_survival': 'mean',
        'survival_percent': lambda x: float(sum(x)) / len(x)
    },
    'Fare': {     # Now work on the "date" column
        'max_fare': 'max',   # Find the max, call the result "max_date"
        'min_fare': 'min',
        'total_of_all_fares': 'sum',
        'fare_range': lambda x: max(x) - min(x)  # Calculate the date range per group
    },
    'Age': ["count", "max"]  # Calculate two results for the 'network' column with a list
}
 
# Perform groupby aggregation by "month", but only on the rows that are of type "call"
df.groupby('Sex').agg(aggregations)
```
